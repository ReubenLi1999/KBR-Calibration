```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
```

```python
idata = np.loadtxt('..//..//output//eq_b_filtered.txt', dtype=np.longdouble)

plt.plot(idata[:])
plt.show()
```

```python

freq_0, psd_0= welch(idata, 0.2, ('kaiser', 30.), idata.__len__(), scaling='density')
plt.loglog(freq_0, np.sqrt(psd_0))
plt.ylabel(r'$ASD(m/\sqrt{Hz})$')
plt.xlabel('Frequency [Hz]')
plt.show()
```
